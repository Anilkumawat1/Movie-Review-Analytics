```python
# Importing the important libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib
sns.set(style="ticks", color_codes=True)
from collections import Counter
```

```python
#Load dataset 
#https://www.kaggle.com/code/robinjrjr/imdb-movie-data-analysis/input
df = pd.read_csv('imdb-movies.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
print('Count of missing values(zeros) in budget_adj colm')
df[df['budget_adj'] == 0].count()
```

```python
df.drop(['imdb_id','budget', 'revenue', 'homepage'], axis=1, inplace=True)
```

```python
df.rename(index=str, columns={"budget_adj": "total_budget", "revenue_adj": "total_revenue"}, inplace=True)
#changing type of date
df['release_date'] = pd.to_datetime(df['release_date'])
df.head(2)
```

```python
df.info()
```

```python
x = df.duplicated().sum()
print("No of Duplicate Values", x)
df.drop_duplicates(inplace=True)
y = df.duplicated().sum()
print("No of Duplicate Values after cleaning", y)
```

```python
df['release_year'].value_counts().head(10).plot.bar(figsize=(12,4))
```

```python
Genres = []
for i in df.genres:
    if type(i) == str:
        for x in i.split("|"):
            Genres.append(x)
```

```python
dicGenresWithCounts = Counter(Genres)
```

```python
genres_datafame = pd.DataFrame.from_dict([dicGenresWithCounts.values()])
genres_datafame.columns=list(dicGenresWithCounts.keys())
```

```python
genres_datafame.head()
```

```python
genres_datafame.plot.bar(figsize=(12,6))
plt.title("No of movies released in each Genres")
plt.xlabel("Genres")
plt.ylabel("No of movies released")
```

```python
df2 = df.copy(deep=True)
df2['profit']=df2.total_revenue-df2.total_budget

top_movies = df2.nlargest(10,'profit')
top_movies.index = top_movies.original_title
top_movies[['original_title','profit']].plot.bar(figsize=(12,4))
```

```python
df.director.value_counts()[:30].plot.pie(autopct='%1.1f%%',figsize=(10,10))
```

```python
# Top 10 Actor/Actress
arrayCast = []
for i in df.cast:
    if type(i) == str:
        for x in i.split('|'):
            arrayCast.append(x)
dicCastWithCounts = Counter(arrayCast)
df_Cast=pd.DataFrame(data={'Cast':list(dicCastWithCounts.keys()),'Count':list(dicCastWithCounts.values())}).sort_values(by='Count',ascending=False)
df_Cast.index = df_Cast.Cast
df_Cast.Count.head(10).plot.bar(figsize=(12,6))
plt.title("Top 10 Actor/Actress")
plt.xlabel("Cast")
plt.ylabel("No of movies released")
```

```python
df1 = df.copy(deep=True)

df1 = df1[df1['total_revenue'] != 0]
df1 = df1[df1['total_budget'] != 0]
#check the data
df1.head(2)
```

```python
#Plot of budget and revenue
sns.pointplot(x="total_budget", y="total_revenue", data=df1)
```

```python
# popularity vs total_revenue

sns.pointplot(x="popularity", y="total_revenue", data=df1);
```

```python
df_q3 = df.copy(deep=True)
df_q3 =df_q3.drop(df_q3.columns.difference(['vote_count','vote_average','release_year']), 1)
votes =(df_q3.groupby('release_year', sort=True).mean())
votes.head(5)
```

```python
votes['vote_count'].plot(title='Average Vote count over the years');
```

```python
votes['vote_average'].plot( title='Average vote over the years');
```
